```python
# Create student class that takes name & marks of 3 subjects as argument in constructor.Then create a method to print the average.
#Non list method
class Student:
    def __init__(self,name,marks1,marks2,marks3):
        self.name=name
        self.marks1=marks1
        self.marks2=marks2
        self.marks3=marks3
    def get_avg(self):
        avg=(self.marks1+self.marks2+self.marks3)/3
        return avg
Student_1=Student("Rafiul Islam",83,87,76)
print(Student_1.name,Student_1.get_avg())
```

```python
class Car:
    def __init__(self):
        self.accelerator = False
        self.brake = False
        self.clutch= False
    def start(self):
        self.clutch = True
        self.accelerator = True                     # The Abstraction method
        self.brake = False
        print("The Car has already started....")
    def stop(self):
        self.clutch = False
        self.accelerator = False
        self.brake = True
        print("The car has stopped....")
car_1 = Car()
car_1.start()
car_1.stop()
```

```python
class Account:
    def __init__(self,balance,account_number):
        self.balance = balance
        self.account_number = account_number
    def debit(self,amount):
        self.balance -= amount
        print(f"{self.account_number} debited {amount} Taka")
        print("Total balance right now:",self.get_balance())                    # The Encapsulation Method
    def credit(self,amount):
        self.balance += amount
        print(f"{self.account_number} credited {amount} Taka")
        print("Total balance right now:",self.get_balance())
    def get_balance(self):
        return self.balance
Account_1 = Account(500,+880167453349)
print(Account_1.balance,Account_1.account_number)
Account_1.debit(100)
Account_1.credit(1000)
```

```python
class Car:
    @staticmethod
    def start():
        accelerator=True
        clutch=True
        Brake=False
        print("The car has already started....")
    @staticmethod
    def stop():                                     # The inheritance method
        accelerator=False
        clutch=False
        Brake=True
        print("The car has stopped....")
class Toyota(Car):
    def __init__(self,name,color):
        self.name=name
        self.color=color
car1=Toyota("Fortuner","Blue")
print(car1.name)
print(car1.start())
```

```python
class Person:
    name = "anonymous"
    @classmethod        # decorator
    def changeName(cls,name):
        cls.name = name         #Class Method
p1 = Person()
p1.changeName("Rafiul islam")
print(p1.name)
```

```python
class Student:
    def __init__(self,math,cse,eee):
        self.math=math
        self.cse=cse
        self.eee=eee
    @property
    def percentage(self):
        return str((self.math+self.cse+self.eee)/3)+"%"
Student_1 = Student(98,97,99)
print(f"Previous result was: {Student_1.percentage}")
Student_1.math = 84
print(f"Current result is: {Student_1.percentage}")
```
