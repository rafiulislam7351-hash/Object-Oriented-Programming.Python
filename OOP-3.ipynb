```python
class Employee:
    company = "IIT"
    def __init__(self, position, salary, **kwargs):
        self.position = position
        self.salary = salary
        super().__init__(**kwargs)

class Programmer:
    def __init__(self, name, language, **kwargs):           #Multiple Inheritance
        self.name = name
        self.language = language
        super().__init__(**kwargs)

class Payment(Employee, Programmer):
    def __init__(self, name, language, position, salary):
        super().__init__(name=name, language=language, position=position, salary=salary)

    def showPayment(self):
        print(f"The month end payment for {self.name} working as a {self.language} "
              f"specialist in {self.position} is {self.salary}$.")

employee1 = Payment("Rafiul Islam", "Python", "Python Developer", 120000)
employee1.showPayment()
```

```python
class TwoDVector:
    def __init__(self,i,j):
        self.i=i
        self.j=j
    def show(self):
        print(f"the vector is {self.i}i+{self.j}j")
class ThreeDVector(TwoDVector):
    def __init__(self,i,j,k):
        super().__init__(i,j)
        self.k=k
    def show3D(self):
            print(f"the vector is {self.i}i+{self.j}j+{self.k}k")
block=TwoDVector(3,4)
block.show()
block_1=ThreeDVector(3,4,5)
block_1.show3D()
```

```python
class Complex:
    def __init__(self,real,imaginary):
        self.real=real
        self.imaginary=imaginary
    def __add__(self,c2):
        sum_real = self.real+c2.real
        sum_imaginary = self.imaginary+c2.imaginary
        return Complex(sum_real,sum_imaginary)
    def __mul__(self, c2):
        new_real = (self.real * c2.real) - (self.imaginary * c2.imaginary)
        new_imaginary = (self.real * c2.imaginary) + (self.imaginary * c2.real)
        return Complex(new_real, new_imaginary)

    def __str__(self):
        sign = "+" if self.imaginary >= 0 else "-"
        return f"{self.real} {sign} {abs(self.imaginary)}i"
c1=Complex(2,3)
c2=Complex(4,5)
print(c1+c2)
print(c1*c2)
```

```python
class Vector:
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    def __add__(self, other):
        result = Vector(self.x+other.x,self.y+other.y,self.z+other.z)
        return result
    def __mul__(self, other):
        result = self.x*other.x+self.y*other.y+self.z*other.z
        return result
    def __str__(self):
        return f"{self.x}i+{self.y}j+{self.z}k"
v1=Vector(1,2,3)
v2=Vector(4,5,6)
print(v1+v2)
print(v1*v2)
```
